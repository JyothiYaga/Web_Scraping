```python
import requests
from bs4 import BeautifulSoup
import csv
import time

serial_number_counter = 0

def find_technews():
    global serial_number_counter

    url = 'https://tv9telugu.com/technology'
    html_text = requests.get(url)
    soup = BeautifulSoup(html_text.content, 'html.parser')

    news_items = soup.find('div', class_='tv9_landingWidget')

    tech_news_data = []

    for news in news_items.find_all('figure'):
        serial_number_counter += 1

        news_title = news.find('div', class_='card_title').text.strip()
        news_link = news.find('a')['href']

        # Check if 'imgThumb' class exists before accessing img tag
        img_container = news.find('div', class_='imgThumb')
        news_imglink = img_container.find('img')['src'] if img_container else "N/A"

        tech_news_data.append({
            'Serial Number': serial_number_counter,
            'Title': news_title,
            'Link': news_link,
            'Image_link': news_imglink
        })

    write_to_csv(tech_news_data)

def write_to_csv(tech_news_data):
    with open('tv9tech_news.csv', 'w', newline='', encoding='utf-8') as csvfile:
        fieldnames = ['Serial Number', 'Title', 'Link', 'Image_link']
        csv_writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

        csv_writer.writeheader()
        csv_writer.writerows(tech_news_data)

if __name__ == '__main__':
    while True:
        find_technews()
        time_wait = 10
        print(f'Waiting {time_wait} minutes...')
        time.sleep(time_wait * 60)
```
